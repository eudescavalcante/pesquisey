```python
import hashlib
import pandas as pd

usuarios = {}
produtos = pd.DataFrame(columns=['Nome do Produto', 'Quantidade', 'Preço', 'Cidade', 'Estado', 'Bairro'])

def cadastrar_usuario():
    cpf = input("Digite o número do CPF: ")
    senha = input("Digite a senha de acesso: ")

    senha = hashlib.sha256(senha.encode()).hexdigest()

    usuarios[cpf] = senha
    print("Usuário cadastrado com sucesso!")

def login():
    cpf = input("Digite o número do CPF: ")
    senha = input("Digite a senha de acesso: ")

    senha = hashlib.sha256(senha.encode()).hexdigest()

    if cpf in usuarios and usuarios[cpf] == senha:
        print("Login realizado com sucesso!")
        return True
    else:
        print("CPF ou senha incorretos.")
        return False

def cadastrar_produto():
    nome_produto = input("Digite o nome do produto: ")
    quantidade = input("Digite a quantidade: ")
    preco = input("Digite o preço: ")
    cidade = input("Digite a cidade: ")
    estado = input("Digite o estado (2 caracteres): ")
    bairro = input("Digite o bairro: ")

    # Adiciona o prefixo R$ ao preço
    preco = "R$" + preco

    produto = pd.Series([nome_produto, quantidade, preco, cidade, estado, bairro], index=produtos.columns)
    produtos.loc[len(produtos)] = produto

    print("Produto cadastrado com sucesso!")

def exibir_produtos():
    if produtos.empty:
        print("Nenhum produto cadastrado.")
    else:
        print(produtos)

def salvar_dados_excel():
    produtos.to_excel("produtos.xlsx", index=False)
    print("Dados salvos no arquivo 'produtos.xlsx'.")

while True:
    print("1 - Cadastrar usuário")
    print("2 - Fazer login")
    print("3 - Cadastrar produto")
    print("4 - Exibir produtos cadastrados")
    print("5 - Salvar dados em Excel")
    print("6 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        cadastrar_usuario()
    elif opcao == "2":
        if login():
            while True:
                print("1 - Cadastrar produto")
                print("2 - Exibir produtos cadastrados")
                print("3 - Fazer logout")

                opcao_logado = input("Escolha uma opção: ")

                if opcao_logado == "1":
                    cadastrar_produto()
                elif opcao_logado == "2":
                    exibir_produtos()
                elif opcao_logado == "3":
                    print("Logout realizado com sucesso!")
                    break
                else:
                    print("Opção inválida.")
        else:
            continue
    elif opcao == "3":
        print("Faça login antes de cadastrar um produto.")
    elif opcao == "4":
        exibir_produtos()
    elif opcao == "5":
        salvar_dados_excel()
    elif opcao == "6":
        break
    else:
        print("Opção inválida.")
```
